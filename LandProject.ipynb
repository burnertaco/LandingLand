```python
import pandas as pd
import folium
df = pd.read_csv('Sold_data.csv')

df.head()
```

```python
# how many rows in df
len(df)
```

```python
# Remove any rows that might be missing location data
df.dropna(subset=['Latitude', 'Longitude'], inplace=True)

# how many rows remain
len(df)
```

```python
# Create a map, centered on the average coordinates of all 9462 locations
my_map = folium.Map(location=[df['Latitude'].mean(), df['Longitude'].mean()], zoom_start=9)

# Loop through your data and add a circle for each point
for idx, row in df.iterrows():
    # Create the text that will appear in the popup
    popup_text = f"Price: ${int(row['Price']):,}<br>Acres: {row['Acres']}"
    
    # Add a blue circle marker to the map
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=1,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        popup=folium.Popup(popup_text, max_width=200)
    ).add_to(my_map)

my_map.save("all_sold.html")
```

```python
# what kind of entries are the days on market (we want numbers, ie ints/floats)
df['Days on Market'].apply(type).value_counts()
```

```python
# filter the DataFrame
# "hot" properties defined as 2-5 acres on market for < 90 days (but more than 0)
acreage_filtered_df = df[(df['Days on Market'] > 0) & (df['Days on Market'] <= 90)].copy()
acreage_filtered_df = acreage_filtered_df[(acreage_filtered_df['Acres'] >= 2) & (acreage_filtered_df['Acres'] <= 10)].copy()

acreage_filtered_df.head()
```

```python
# Create a new map, centered on the average coordinates of your filtered data
map_of_hot_properties = folium.Map(location=[acreage_filtered_df['Latitude'].mean(), acreage_filtered_df['Longitude'].mean()], zoom_start=9)

# Loop through your filtered data and add a green circle for each property
for idx, row in acreage_filtered_df.iterrows():
    # Create the popup text, including Days on Market for verification
    popup_text = f"Price: ${int(row['Price']):,}<br>" \
                 f"Acres: {row['Acres']}<br>" \
                 f"Days on Market: {int(row['Days on Market'])}"
    
    # Add a green circle marker to the map to signify it's a "hot" property
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=1,
        color='green',
        fill=True,
        fill_color='#28a745',
        popup=folium.Popup(popup_text, max_width=200)
    ).add_to(map_of_hot_properties)

map_of_hot_properties.save("hot_pockets.html")
```

```python
import numpy as np
from sklearn.cluster import DBSCAN
import matplotlib.colors as mcolors

# convert to numpy array
coords = acreage_filtered_df[['Latitude', 'Longitude']].to_numpy()
coords_in_radians = np.radians(coords)

# > 15 properties within a 10-mile radius
radius = 10
samples = 15
dbscan = DBSCAN(
    eps=(radius/3959), 
    min_samples=samples, 
    algorithm='ball_tree', 
    metric='haversine')

# run the scan
cluster_labels = dbscan.fit_predict(coords_in_radians)
# assigned a cluster (-1 for noise)
acreage_filtered_df['cluster'] = cluster_labels

# Create the map with the new clusters
final_map = folium.Map(location=[acreage_filtered_df['Latitude'].mean(), acreage_filtered_df['Longitude'].mean()], zoom_start=9)
unique_clusters = acreage_filtered_df['cluster'][acreage_filtered_df['cluster'] != -1].unique()
colors = list(mcolors.TABLEAU_COLORS.values()) * 10

for cluster_id in unique_clusters:
    cluster_points = acreage_filtered_df[acreage_filtered_df['cluster'] == cluster_id]
    center_lat = cluster_points['Latitude'].mean()
    center_lon = cluster_points['Longitude'].mean()
    
    folium.Circle(
        location=[center_lat, center_lon],
        radius=radius * 1609,        # miles in meters
        color=colors[cluster_id],
        fill=True,
        fill_opacity=0.2,
        popup=f"Hot Pocket #{cluster_id}<br>{len(cluster_points)} properties"
    ).add_to(final_map)
    
    for idx, row in cluster_points.iterrows():
        folium.CircleMarker(
            location=[row['Latitude'], row['Longitude']],
            radius=1,
            color=colors[cluster_id],
            fill=True,
            fill_color=colors[cluster_id],
            fill_opacity=0.8
        ).add_to(final_map)

# set the rigth zoom
final_map.fit_bounds(final_map.get_bounds())
final_map.save("clusters.html")
```

```python
# tennessee df
tennessee_df = acreage_filtered_df[
    (acreage_filtered_df['cluster'] != -1) &
    (acreage_filtered_df['State'] == 'Tennessee')
].copy()

# two different df by above/below 5 acres
tennessee_df_lt_5 = tennessee_df[tennessee_df['Acres'] <= 5].copy()
tennessee_df_gt_5 = tennessee_df[tennessee_df['Acres'] > 5].copy()

# check lengths
len(tennessee_df), len(tennessee_df_lt_5), len(tennessee_df_gt_5)
```

```python
# alabama df
alabama_df = acreage_filtered_df[
    (acreage_filtered_df['cluster'] != -1) &
    (acreage_filtered_df['State'] == 'Alabama')
].copy()

# two different df by above/below 5 acres
alabama_df_lt_5 = alabama_df[(alabama_df['Acres'] <= 5)].copy()
alabama_df_gt_5 = alabama_df[(alabama_df['Acres'] > 5)].copy()

# check lengths
len(alabama_df), len(alabama_df_lt_5), len(alabama_df_gt_5)
```
